```python
import pandas as pd
# Load the dataset
file_path = r"C:\Users\vange\Desktop\github project 2\Sleep_health_and_lifestyle_dataset.csv"
df = pd.read_csv(file_path)

# Display the first few rows
df.head()
```

```python
df.info()
```

```python
# Ensure 'Quality of Sleep' does not exceed 'Sleep Duration'
df['Effective Sleep'] = df[['Quality of Sleep', 'Sleep Duration']].min(axis=1)

# Calculate Sleep Efficiency as (Effective Sleep / Sleep Duration) * 100
df['Sleep Efficiency (%)'] = (df['Effective Sleep'] / df['Sleep Duration']) * 100

# Drop the temporary 'Effective Sleep' column (optional)
df.drop(columns=['Effective Sleep'], inplace=True)

# Display the first few rows to confirm
print(df.head())

# Check for any efficiency values greater than 100% (should be none now)
assert (df['Sleep Efficiency (%)'] <= 100).all(), "There are still values above 100%!"

print("Sleep Efficiency KPI successfully calculated.")
```

```python
# Define a function to classify blood pressure
def classify_blood_pressure(row):
    if row['Systolic Blood Pressure'] < 90 or row['Diastolic Blood Pressure'] < 60:
        return 'Hypotensive'  # Low Blood Pressure
    else:
        return 'Normal/Hypertensive'  # Normal or High Blood Pressure

# Apply the function to create a new column
df['Blood Pressure Classification'] = df.apply(classify_blood_pressure, axis=1)

# Display the first few rows to confirm
df.head()
```
